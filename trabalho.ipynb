```python
import random
import time
import matplotlib.pyplot as plt
```

Essas linhas importam as bibliotecas necessárias para o código. A biblioteca random é usada para gerar números aleatórios, time é utilizada para medir o tempo de execução dos algoritmos e matplotlib.pyplot é usada para criar gráficos.

```python
def pivo(vet, baixo, alto):
    i = baixo - 1
    pivo = vet[alto]

    for j in range(baixo, alto):
        if vet[j] <= pivo:
            i += 1
            vet[i], vet[j] = vet[j], vet[i]

    vet[i + 1], vet[alto] = vet[alto], vet[i + 1]
    return i + 1
```

Esta função pivo é responsável por realizar a etapa de particionamento do algoritmo Quick Sort. Ela recebe um vetor (vet), um índice inferior (baixo) e um índice superior (alto). A função seleciona o elemento pivô, que é o último elemento do vetor (vet[alto]), e percorre o vetor comparando cada elemento com o pivô. Os elementos menores ou iguais ao pivô são colocados à esquerda dele, e os elementos maiores são colocados à direita. No final, o pivô é colocado na posição correta e a função retorna essa posição.

```python
def quick(vet, baixo, alto):
    while baixo < alto:
        indice_pivo = pivo(vet, baixo, alto)
        if indice_pivo - baixo < alto - indice_pivo:
            quick(vet, baixo, indice_pivo - 1)
            baixo = indice_pivo + 1
        else:
            quick(vet, indice_pivo + 1, alto)
            alto = indice_pivo - 1
```

A função quick implementa o algoritmo Quick Sort. Ela recebe um vetor (vet), um índice inferior (baixo) e um índice superior (alto). O algoritmo usa a função pivo para obter o índice do pivô e, em seguida, recursivamente chama a função quick para ordenar as duas partições do vetor, uma à esquerda e outra à direita do pivô. A função é executada em um loop enquanto o índice inferior (baixo) é menor que o índice superior (alto)

```python
def insertion(vet):
    tamanho = len(vet)

    for i in range(1, tamanho):
        chave = vet[i]
        j = i - 1

        while j >= 0 and vet[j] > chave:
            vet[j + 1] = vet[j]
            j -= 1

        vet[j + 1] = chave
```

A função insertion implementa o algoritmo Insertion Sort. Ela recebe um vetor (vet) e ordena os elementos do vetor em ordem crescente. O algoritmo percorre o vetor a partir do segundo elemento, comparando-o com os elementos anteriores e movendo-os para a direita até encontrar a posição correta para inserir o elemento atual.

```python
tamanhos = [50, 500, 5000, 50000] 
TimeQScresce = []
TimeQSdecresce = []
TimeQSrandom = []
TimeIScresce = []
TimeISdecresce = []
TimeISrandom = []
```

Essa parte do código cria listas vazias para armazenar os tempos de execução do Quick Sort (TimeQScresce, TimeQSdecresce, TimeQSrandom) e do Insertion Sort (TimeIScresce, TimeISdecresce, TimeISrandom) para diferentes tamanhos de vetor (tamanhos).

```python
for tam in tamanhos:
    VcresceQS = list(range(tam))
    VcresceIS = list(range(tam))
    start = time.time()
    quick(VcresceQS, 0, len(VcresceQS) - 1)
    finish = time.time()
    Tempo_gasto = finish - start
    TimeQScresce.append(Tempo_gasto)

    start = time.time()
    insertion(VcresceIS)
    finish = time.time()
    Tempo_gasto = finish - start
    TimeIScresce.append(Tempo_gasto)

    VdecresceQS = list(range(tam, 0, -1))
    VdecresceIS = list(range(tam, 0, -1))
    start = time.time()
    quick(VdecresceQS, 0, len(VdecresceQS) - 1)
    finish = time.time()
    Tempo_gasto = finish - start
    TimeQSdecresce.append(Tempo_gasto)

    start = time.time()
    insertion(VdecresceIS)
    finish = time.time()
    Tempo_gasto = finish - start
    TimeISdecresce.append(Tempo_gasto)
    
    VrandomQS = random.sample(range(tam), tam)
    VrandomIS = random.sample(range(tam), tam)
    start = time.time()
    quick(VrandomQS, 0, len(VrandomQS) - 1)
    finish = time.time()
    Tempo_gasto = finish - start
    TimeQSrandom.append(Tempo_gasto)

    start = time.time()
    insertion(VrandomIS)
    finish = time.time()
    Tempo_gasto = finish - start
    TimeISrandom.append(Tempo_gasto)
```

Nesse loop, para cada tamanho de vetor em tamanhos, são executados os seguintes passos:

Criação de vetores: São criados três tipos diferentes de vetores: um vetor crescente (VcresceQS e VcresceIS), um vetor decrescente (VdecresceQS e VdecresceIS) e um vetor aleatório (VrandomQS e VrandomIS).

Medição do tempo de execução do Quick Sort: O tempo de início (start) é registrado, o algoritmo Quick Sort é chamado para o vetor correspondente e, em seguida, o tempo de término (finish) é registrado. A diferença entre o tempo de término e o tempo de início é calculada e armazenada na lista TimeQScresce, TimeQSdecresce ou TimeQSrandom, dependendo do tipo de vetor.

Medição do tempo de execução do Insertion Sort: O processo é semelhante ao anterior, mas agora o algoritmo Insertion Sort é usado e os tempos de execução são armazenados nas listas TimeIScresce, TimeISdecresce ou TimeISrandom.

```python
plt.plot(tamanhos, TimeQScresce, label='Quick Sort', marker='.', color='red')
plt.plot(tamanhos, TimeIScresce, label='Insertion Sort', marker='.', color='green')
plt.ylabel('Tempo de processamento (S)')
plt.xlabel('Tamanho do vetor')
plt.title('Comparação - Vetor Crescente')
plt.legend()
plt.show()
```

Essas linhas criam um gráfico comparativo para o caso em que os vetores estão em ordem crescente. Os tempos de execução do Quick Sort e do Insertion Sort são plotados em relação aos tamanhos dos vetores. O eixo y representa o tempo de execução em segundos, o eixo x representa o tamanho do vetor e o título e legenda são adicionados para identificar os algoritmos. O gráfico é exibido com o comando plt.show().

```python
plt.plot(tamanhos, TimeQSdecresce, label='Quick Sort', marker='.', color='red')
plt.plot(tamanhos, TimeISdecresce, label='Insertion Sort', marker='.', color='green')
plt.ylabel('Tempo de processamento (S)')
plt.xlabel('Tamanho do vetor')
plt.title('Comparação - Vetor Decrescente')
plt.legend()
plt.show()
```

Essas linhas criam um gráfico comparativo para o caso em que os vetores estão em ordem decrescente. O processo é semelhante ao anterior, mas os tempos de execução correspondem aos vetores decrescentes.

```python
plt.plot(tamanhos, TimeQSrandom, label='Quick Sort', marker='.', color='red')
plt.plot(tamanhos, TimeISrandom, label='Insertion Sort', marker='.', color='green')
plt.ylabel('Tempo de processamento (S)')
plt.xlabel('Tamanho do vetor')
plt.title('Comparação - Vetor Aleatório')
plt.legend()
plt.show()
```

Essas linhas criam um gráfico comparativo para o caso em que os vetores são aleatórios. O processo é semelhante aos anteriores, mas os tempos de execução correspondem aos vetores aleatórios.
